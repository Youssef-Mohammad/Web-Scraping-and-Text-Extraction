```python
import requests
import spacy

from bs4 import BeautifulSoup
import re
```

```python
url = "https://en.wikipedia.org/wiki/Napoleon"
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
# Getting text in paragraphs and headings only
pattern = r'<p>(.*?)<\/p>|<h\d>(.*?)<\/h\d>'
```

```python
matches = re.findall(pattern, response.text, re.DOTALL)
```

```python
captured_text = ""

for match in matches:
    for word in match:
        if word:
            captured_text += word + "\n"
```

```python
captured_text
```

```python
soup = BeautifulSoup(captured_text, 'html.parser')
soup
```

```python
text = soup.get_text()
len(text)
```

```python
# Remove extra whitespace
text = re.sub(r'\s+', ' ', text)
len(text)
```

```python
# Remove Parentheses
text = re.sub(r'\[', ' ', text)
text = re.sub(r'\]', ' ', text)
text = re.sub(r'\(', ' ', text)
text = re.sub(r'\)', ' ', text)
text = re.sub(r'\.', ' ', text)
text = re.sub(r'\-', ' ', text)
text = re.sub(r'\,', ' ', text)
text = re.sub(r'  ', ' ', text)

len(text)
```

```python
text
```

```python
# Remove leading and trailing whitespace
text = text.strip()
len(text)
```

```python
text
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
```

```python
# Removing Stop Words
text = ' '.join([token.text for token in doc if not token.is_stop])
len(text)
```

```python
doc = nlp(text)
```

```python
lemmas=[]

for token in doc:
  lemmas.append(token.lemma_)
  print(f"{token} | {token.lemma_}")
```

```python
lemmas
```

```python
unique_lemmas = set(lemmas)
unique_lemmas
```
